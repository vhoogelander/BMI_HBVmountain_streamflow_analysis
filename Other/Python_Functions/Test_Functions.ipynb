```python
 
def hillslopeHRU(Precipitation, Temperature, ETP, storages, parameters):
    
    Snow = 0
    Interception = 0
    Snow_Cover = 0
    Total_Melt_Glacier = 0
    Total_Effective_Precipitation = 0
    Total_Interception_Evaporation = 0
    
    # calculate the model components that are height dependend

    # snow component
    Melt_Glacier, Melt, Snow = snow(0, Precipitation, Temperature, float(storages.Snow), float(parameters.Meltfactor), float(parameters.Mm), float(parameters.Temp_Thresh))
    #interception component
    Effective_Precipitation, Interception_Evaporation, Interception = interception(ETP, Precipitation, Temperature, float(storages.Interception), float(parameters.Interceptionstoragecapacity), float(parameters.Temp_Thresh))
    # the melt, effective precipitation and evaporation can be summed up over all elevations according to the areal extent
    Total_Effective_Precipitation += (Effective_Precipitation + Melt) 
    #global Total_Melt += (Melt * Area_Elevations[i])
    Total_Interception_Evaporation += (Interception_Evaporation)
    Total_Melt_Glacier += (Melt_Glacier)
        # get the extend of snow cover, if there is snow in the snow bucket the snow cover is 1
    if Snow > 1:
        Snow_Cover = 1          
    #soil storage component
    Overlandflow, Preferentialflow, Soil_Evaporation, Soilstorage = soilstorage(Total_Effective_Precipitation, Total_Interception_Evaporation, ETP, float(storages.Soil), float(parameters.beta), float(parameters.Ce), float(parameters.Ratio_Pref), float(parameters.Soilstoragecapacity))
    GWflow = Preferentialflow
    #fast storage
    Fast_Discharge, Faststorage = faststorage(Overlandflow, float(storages.Fast), float(parameters.Kf))

    #calculate total outflow
    Flows = Fast_Discharge + GWflow + Total_Interception_Evaporation + Soil_Evaporation
#     assert hill.Potential_Evaporation_Mean - Soil_Evaporation - hill.Total_Interception_Evaporation >= - 5 * eps
    # change discharges and evaporation fluxed according to areal fraction
    Fast_Discharge = Fast_Discharge 
    GWflow = GWflow 
    Total_Interception_Evaporation = Total_Interception_Evaporation
    Soil_Evaporation = Soil_Evaporation
    # returning all fluxes (evporative, discharge)
    hill_out = [Fast_Discharge, GWflow, Soil_Evaporation, Total_Interception_Evaporation] #Outflows(Fast_Discharge, GWflow, Soil_Evaporation, Total_Interception_Evaporation)
    #returning all storages
    hill_storages = [Faststorage, Interception, Snow, Snow_Cover, Soilstorage]#Storages(Faststorage, Interception, Snow, Snow_Cover, Soilstorage)
    # total interception should be zero again for next run
    Total_Interception_Evaporation = 0
    Total_Effective_Precipitation = 0

    #assertion water balance
    Prec = 0
    Interception_Storage_New = 0
    Snow_Storage_New = 0
    Interception_Storage_Old = 0
    Snow_Storage_Old = 0
    # get total storage of interception and snow over all elevation zones
   
    Prec += Precipitation
    Interception_Storage_New += hill_storages[1]
    Snow_Storage_New += hill_storages[2]
    Interception_Storage_Old += storages.Interception
    Snow_Storage_Old += storages.Snow 
    
    Flows_Area = Fast_Discharge + GWflow + Total_Interception_Evaporation + Soil_Evaporation
    All_Storages = (hill_storages[0] - storages.Fast) + (hill_storages[4] - storages.Soil) + (Snow_Storage_New -Snow_Storage_Old) + (Interception_Storage_New - Interception_Storage_Old)


    
    return hill_out, hill_storages, Precipitation, All_Storages
```

```python
def riparianHRU(Precipitation, Temperature, ETP, Riparian_Discharge, storages, parameters):
    # Are_Elevations gives the areal percentage of each elevation band. The sum has to be 1
    # Area_elevations, Precipitation, Temp_elevation, Snowstorage, Interceptionstorage has to be array of length Nr_Elevationbands
    # riparian HRU has no preferential flow

#     assert rip.Total_Interception_Evaporation == 0
#     assert rip.Total_Effective_Precipitation == 0
#     assert 0.9999999 <= sum(rip.Area_Elevations) <= 1.000001 or sum(rip.Area_Elevations) == 0
#     assert rip.Area_HRU >= 0 and rip.Area_HRU <= 1
#     assert rip.Nr_Elevationbands >= 1
    
    Snow = 0
    Interception = 0
    Snow_Cover = 0
    Total_Effective_Precipitation = 0
    Total_Interception_Evaporation = 0
    # snow component
#     assert rip.Area_Glacier == 0.0
    Melt_Glacier, Melt, Snow = snow(0, Precipitation, Temperature, float(storages.Snow), float(parameters.Meltfactor), float(parameters.Mm), float(parameters.Temp_Thresh))
#     assert Melt_Glacier == 0.0
    #interception component
    Effective_Precipitation, Interception_Evaporation, Interception = interception(ETP, Precipitation, Temperature, float(storages.Interception), float(parameters.Interceptionstoragecapacity), float(parameters.Temp_Thresh))
    # the melt, effective precipitation and evaporation can be summed up over all elevations according to the areal extent
    Total_Effective_Precipitation += (Effective_Precipitation + Melt) 
    #global Total_Melt += (Melt * Area_Elevations[i])
    Total_Interception_Evaporation += Interception_Evaporation 
    #get snow cover extent
    if Snow > 1:
        Snow_Cover = 1


    #soil storage component
    Overlandflow, Soil_Evaporation, Soilstorage = ripariansoilstorage(Total_Effective_Precipitation, Total_Interception_Evaporation, ETP, Riparian_Discharge, float(storages.Soil), float(parameters.beta), float(parameters.Ce), float(parameters.Drainagecapacity), float(parameters.Soilstoragecapacity))
    #fast storage
    Fast_Discharge, Faststorage = faststorage(Overlandflow, float(storages.Fast), float(parameters.Kf))
    GWflow = 0
    #calculate total outflow
    Flows = Fast_Discharge + GWflow + Total_Interception_Evaporation + Soil_Evaporation
#     assert rip.Potential_Evaporation_Mean - Soil_Evaporation - rip.Total_Interception_Evaporation >= - 5 * eps
    # change discharges according to areal fraction
    Fast_Discharge = Fast_Discharge
    Total_Interception_Evaporation = Total_Interception_Evaporation 
    Soil_Evaporation = Soil_Evaporation 
    # return water flows, evaporation, and states of the storage components
    rip_out =  [Fast_Discharge, GWflow, Soil_Evaporation, Total_Interception_Evaporation]#Outflows(Fast_Discharge, GWflow, Soil_Evaporation, Total_Interception_Evaporation)
    rip_storages = [Faststorage, Interception, Snow, Snow_Cover, Soilstorage] #Storages(Faststorage, Interception, Snow, Snow_Cover, Soilstorage)

    Total_Interception_Evaporation = 0
    Total_Effective_Precipitation = 0

    Prec = 0
    Interception_Storage_New = 0
    Snow_Storage_New = 0
    Interception_Storage_Old = 0
    Snow_Storage_Old = 0
    
    Prec += Precipitation
    Interception_Storage_New += rip_storages[1] 
    Snow_Storage_New += rip_storages[2]
    Interception_Storage_Old += storages.Interception
    Snow_Storage_Old += storages.Snow
   

    Flows_Area = Fast_Discharge + GWflow + Total_Interception_Evaporation + Soil_Evaporation
    All_Storages = (rip_storages[0] - storages.Fast) + (rip_storages[4] -storages.Soil) + (Snow_Storage_New -Snow_Storage_Old) + (Interception_Storage_New - Interception_Storage_Old)


    return rip_out, rip_storages, Precipitation, All_Storages

```

```python
def allHRU(Precipitation, Temperature, ETP, Riparian_Discharge,
           bare_storage, forest_storage, grass_storage, rip_storage,
           bare_parameters, forest_parameters, grass_parameters, rip_parameters,
           Slowstorage, slow_parameters):
    #this function runs thy model for different HRUs
    #bare rock HRU
    bare_outflow, bare_storage, Bare_precipitation, Bare_Storages = hillslopeHRU(Precipitation, Temperature, ETP, bare_storage, bare_parameters)
    # forest HRU
    forest_outflow, forest_storage, Forest_precipitation, Forest_Storages = hillslopeHRU(Precipitation, Temperature, ETP, forest_storage, forest_parameters)
    # Grassland HRU
    grass_outflow, grass_storage, Grass_precipitation, Grass_Storages= hillslopeHRU(Precipitation, Temperature, ETP, grass_storage, grass_parameters)
    # riparian HRU
    rip_outflow, rip_storage, Rip_precipitation, Rip_Storages= riparianHRU(Precipitation, Temperature, ETP, Riparian_Discharge, rip_storage, rip_parameters)
    # total flow into groundwater is the weighted sum of the HRUs according to areal extent (this was already done in hillslopeHRU)
    Total_GWflow = bare_outflow[1] + forest_outflow[1] + grass_outflow[1]
    # Groundwater storage
    Riparian_Discharge, Slow_Discharge, Slowstorage_New = slowstorage(Total_GWflow, Slowstorage, 1, slow_parameters.Ks, slow_parameters.Ratio_Riparian)
    #return all storage values, all evaporation values, Fast_Discharge and Slow_Discharge
    # calculate total discharge of the timestep using weighted sum of each HRU
    Total_Discharge = bare_outflow[0]  + forest_outflow[0] + grass_outflow[0]  + rip_outflow[0]  + Slow_Discharge
    Total_Soil_Evaporation = bare_outflow[2] + forest_outflow[2] + grass_outflow[2]  + rip_outflow[2]
    Total_Interception_Evaporation = bare_outflow[3]  + forest_outflow[3] + grass_outflow[3] + rip_outflow[3]
    Total_Flows = Total_Discharge + Total_Soil_Evaporation + Total_Interception_Evaporation + Riparian_Discharge
    Total_Storages = Bare_Storages + Forest_Storages  + Grass_Storages  + Rip_Storages  + Slowstorage_New - Slowstorage
    Precipitation = Bare_precipitation + Forest_precipitation  + Grass_precipitation  + Rip_precipitation 
    #@assert -0.00000001 <= Precipitation + rip_input.Riparian_Discharge - (Total_Flows + Total_Storages) <= 0.00000001
    Waterbalance = Precipitation + Riparian_Discharge - (Total_Flows + Total_Storages)
    
    initial_storage= np.zeros([1, 5])
    name_storage = ('Fast', 'Interception', 'Snow', 'Snow_Cover', 'Soil')
    df_bare_storage = df_forest_storage = df_grass_storage = df_rip_storage = pd.DataFrame(initial_storage, columns=name_storage) 
    
    for i in range(len(in_bare)):
        df_bare_storage.iloc[:, i] = bare_storage[i]
        df_forest_storage.iloc[:, i] = forest_storage[i]
        df_grass_storage.iloc[:, i] = grass_storage[i]
        df_rip_storage.iloc[:, i] = rip_storage[i]
    
    bare_storage = df_bare_storage
    forest_storage = df_forest_storage
    grass_storage = df_grass_storage
    rip_storage = df_rip_storage
    
    return Riparian_Discharge, Total_Discharge, Total_Interception_Evaporation, Total_Soil_Evaporation, bare_storage, forest_storage, grass_storage, rip_storage, Slowstorage_New, Waterbalance, Precipitation
# pd.DataFrame(bare_storage, columns=bare_parameters)
```

```python
def run_model(Precipitation, Temperature, Evaporation_Mean, Riparian_Discharge,
                bare_storage, forest_storage, grass_storage, rip_storage, Slowstorage,
                bare_parameters, forest_parameters, grass_parameters, rip_parameters, slow_parameters):
    # the function takes as input the parameters of each HRU, the inital storage values of each HRU, the inital value of the slow storage
    # KS, ratio riparian, all inputs
    
    #define maximum time
    tmax = len(Precipitation)
    
    Discharge = np.zeros(tmax)
    WBtotal = np.zeros(tmax)
    Snow_Extend = np.zeros(tmax)
    
    # store the initial storage values, Does not take into account GW storage!!
    # Assumption_ GW storage is 0 at start
    Initial_Storage_bare = sum(bare_storage.Fast) + sum(bare_storage.Interception) + sum(bare_storage.Snow) + sum(bare_storage.Soil)
    Initial_Storage_forest = forest_storage.Fast + sum(forest_storage.Interception) + sum(forest_storage.Snow) + sum(forest_storage.Soil)
    Initial_Storage_grass = sum(grass_storage.Fast) + sum(grass_storage.Interception) + sum(grass_storage.Snow) + sum(grass_storage.Soil)
    Initial_Storage_rip = sum(rip_storage.Fast) + sum(rip_storage.Interception) + sum(rip_storage.Snow) + sum(rip_storage.Soil)
    Initial_Storage = Initial_Storage_bare + Initial_Storage_forest + Initial_Storage_grass + Initial_Storage_rip + Slowstorage
    
    for t in range(tmax):
        # at each timestep new temp, precipitation and Epot values have to be delivered
        # areas don't change
        # riparian discharge from former timestep has to be used
        # gives the current precipitation, evaporation and temperature
        Evaporation_Mean_Current = Evaporation_Mean[t]
        Precipitation_Current = Precipitation[t]
        Temperature_Current = Temperature[t] 
        
#         bare_input = input_timestep(bare_input, Evaporation_Mean_Current, Precipitation_Current, Temperature_Current)
#         forest_input = input_timestep(forest_input, Evaporation_Mean_Current, Precipitation_Current, Temperature_Current)
#         grass_input = input_timestep(grass_input, Evaporation_Mean_Current, Precipitation_Current, Temperature_Current)
#         rip_input = input_timestep(rip_input, Evaporation_Mean_Current, Precipitation_Current, Temperature_Current)
        
        Riparian_Discharge, Total_Discharge, Total_Interception_Evaporation, Total_Soil_Evaporation, bare_storage, forest_storage, grass_storage, rip_storage, Slowstorage, WB, Total_Prec = allHRU(Precipitation_Current, Temperature_Current,  Evaporation_Mean_Current, Riparian_Discharge,
           bare_storage, forest_storage, grass_storage, rip_storage,
           bare_parameters, forest_parameters, grass_parameters, rip_parameters,
           Slowstorage, slow_parameters)
        Riparian_Discharge, Total_Discharge, Total_Interception_Evaporation, Total_Soil_Evaporation, bare_storage, forest_storage, grass_storage, rip_storage, Slowstorage, WB, Total_Prec = float(Riparian_Discharge), float(Total_Discharge), float(Total_Interception_Evaporation), float(Total_Soil_Evaporation), bare_storage, forest_storage, grass_storage, rip_storage, Slowstorage, WB, Total_Prec
        # give new riparian discharge as input for next timestep
        Riparian_Discharge = Riparian_Discharge
        # save the fluxes of the current timestep
        Discharge[t] = Total_Discharge # mm convert it to meter and than * area / seconds in one day
        # OPTIONAL: store the total amount of snow in each elevation (considering whole catchment)
#         Snow_Elevations, Snow_Extend[t] = snowperelevation(bare_input, forest_input, grass_input, rip_input, bare_storage, forest_storage, grass_storage, rip_storage, Total_Elevationbands)
        #Bare_Snow[t,:]::Array{Float64,1} = bare_storage.Snow
        

        WBtotal[t] = WB
                
    # Check Water Balance
    Waterbalance = sum(WBtotal)
#     assert Waterbalance <= 10**(-10)
    
    return Discharge, Snow_Extend, Waterbalance
```
