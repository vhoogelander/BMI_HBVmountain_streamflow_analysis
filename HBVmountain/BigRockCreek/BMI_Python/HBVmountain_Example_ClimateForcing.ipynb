## Historical climate data

```python
calibration = pd.read_csv('Data/BigCreek/calibration_observation.csv', index_col=[0])
calibration.index = pd.to_datetime(calibration.index)

path = 'Data/BigCreek/HBVmountain_GFDL-CM4_BigCreek_1975_2005.nc'
ds = nc.Dataset(path)

prec = (ds['pr'][:])*86400
temp = (ds['tas'][:]) -273

preclist = []
tlist = []

for i in range(len(temp)):
    preclist.append(prec[i])
    tlist.append(temp[i])

base = datetime.datetime.strptime('1975-01-01', '%Y-%m-%d')
date_list = [(base + datetime.timedelta(days=x)).date() for x in range(int(ds['time'][:][-1] - ds['time'][:][0] + 1))]

hist_forcing = pd.DataFrame({'prec':preclist, 'temp':tlist}, index=date_list)

start_date = pd.to_datetime('1975-01-01').date()
end_date = pd.to_datetime('2004-12-31').date()
# Select DataFrame rows between two dates
mask = (hist_forcing.index >= start_date) & (hist_forcing.index <= end_date)
hist_forcing = hist_forcing.loc[mask]
```

## Future climate forcing

```python
pathssp245 = 'Data/BigCreek/HBVmountain_GFDL-CM4_BigCreek_ssp245_2050_2100.nc'
ds_ssp245 = nc.Dataset(pathssp245)

prec_ssp245 = (ds_ssp245['pr'][:])*86400
temp_ssp245 = (ds_ssp245['tas'][:]) -273

preclist_ssp245 = []
tlist_ssp245 = []

for i in range(len(temp_ssp245)):
    preclist_ssp245.append(prec_ssp245[i])
    tlist_ssp245.append(temp_ssp245[i])
    
pathssp585 = 'Data/BigCreek/HBVmountain_GFDL-CM4_BigCreek_ssp585_2050_2100.nc'
ds_ssp585 = nc.Dataset(pathssp585)

prec_ssp585 = (ds_ssp585['pr'][:])*86400
temp_ssp585 = (ds_ssp585['tas'][:]) -273

preclist_ssp585 = []
tlist_ssp585 = []
for i in range(len(temp_ssp585)):
    preclist_ssp585.append(prec_ssp585[i])
    tlist_ssp585.append(temp_ssp585[i])
    
    
base = datetime.datetime.strptime('2050-01-01', '%Y-%m-%d')
date_list = [(base + datetime.timedelta(days=x)).date() for x in range(int(ds_ssp585['time'][:][-1] - ds_ssp585['time'][:][0] + 1))]

future_forcing = pd.DataFrame({'prec_ssp245':preclist_ssp245, 'temp_ssp245':tlist_ssp245, 'prec_ssp585':preclist_ssp585, 'temp_ssp585':tlist_ssp585}, index=date_list)


start_date = pd.to_datetime('2070-01-01').date()
end_date = pd.to_datetime('2099-12-31').date()
# Select DataFrame rows between two dates
mask = (future_forcing.index >= start_date) & (future_forcing.index <= end_date)
future_forcing = future_forcing.loc[mask]
```

#### Observations

```python
rolling_obs = forcing['streamflow'].rolling(window=7).mean()
#     df['MA'] = df.rolling(window=5).mean()
doy_obs = []
magn_obs = []
for i in range(1989, 2005):
    rolling_year_obs = pd.DataFrame(rolling_obs.loc[(rolling_obs.index >= datetime.datetime(i, 3, 1)) & (rolling_obs.index < datetime.datetime(i, 11, 1))])
    rolling_year_obs.reset_index(inplace=True)
    rolling_year_obs.rename(columns={"index":"Date"}, inplace=True)
    rolling_year_obs['cs'] = rolling_year_obs['streamflow'].cumsum()
    doy_obs.append(rolling_year_obs.loc[rolling_year_obs.cs >= (0.5*rolling_year_obs.cs.max())].index[0])
#     magn_hist.append(rolling_year_hist.cs.min())
    magn_obs.append(rolling_year_obs.loc[rolling_year_obs.cs >= (0.5*rolling_year_obs.cs.max())].cs.iloc[-1])
```
